## Outlier Analysis

```python
# plot boxplot to visualize
%matplotlib inline
```

```python
plt.boxplot(x['var_1'])
```

```python
cnames=x.columns
```

```python
cnames
```

```python
# detect and replace  outliers with NAN  from the training data 
for i in cnames:
    print(i)
    q75, q25 =np.percentile(x.loc[:,i],[75,25])
    iqr= q75-q25
    
    min= q25-(iqr*1.5)
    max= q75+ (iqr*1.5)
    
    print(min)
    print(max)
    
    ## replacing with NAs
    x.loc[x[i]< min,:i]=np.nan
    x.loc[x[i]> max,:i]=np.nan
```

```python
## Caluclate missing value
missing_val=pd.DataFrame(x.isnull().sum())
```

```python
missing_val
```

```python
## impute with mean
x_new=x.apply(lambda x: x.fillna(x.mean()),axis=0)
```
